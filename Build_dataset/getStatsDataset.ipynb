```python
final_datasets = {}

for k,v in datasets.items():

    lang = k
    texts_cleaned = v
    count = 0
    count2 = 0
    unique_sentences = dict()
    final_datasets[lang] = []
    file_final = open("scratch/Build_dataset_wiki/final_dataset/" + lang + ".txt","w")
    print(len(texts_cleaned))
    for text in texts_cleaned:
        sent = (" ".join(text["token"]))
        entity_pair = (text["h"]["name"],text["t"]["name"])
        if sent in unique_sentences:
            if(entity_pair in unique_sentences[sent]):
                continue
            elif(len(unique_sentences[sent]) > 2):
                continue
            else:
                unique_sentences[sent].append(entity_pair)
        else:
            unique_sentences[sent] = [entity_pair]
        
        if text["relation"] not in rel_dict:
            text["relation"] = 'NA'
        elif not text["relation"] == 'NA':
            count += 1
        final_datasets[lang].append(text)
        str_temp = json.dumps(text,ensure_ascii = False)
        count2 += 1
        file_final.write(str_temp + "\n")
        
    file_final.close()
        
    print("For language {}, out of {} sentences, {} were had non-NA relations".format(lang,count2,count))
        

for k,v in final_datasets.items():
    
    unique_sentences = set()
    
    lang = k
    texts_cleaned = v
    for text in texts_cleaned:
        unique_sentences.add((text["h"]["name"],text["t"]["name"]))
        
    print("For language {}, there were a total of {} sentences and {} instances of facts".format(lang,len(texts_cleaned),len(unique_sentences)))
    
```
